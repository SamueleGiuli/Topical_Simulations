```python
import numpy as np
import matplotlib.pyplot as plt
```

PARAMETRI:

```python
mode = ["M1","M2"]
alpha = ["a01", "a02", "a03", "a05", "a1", "a2", "a5", "a10"]
eps = ["eps001", "eps01", "eps05", "eps1"]

moden = ["m=1, ", "m=2, "]
alphan= ["a=0.1", "a=0.2","a=0.3","a=0.5","a=1.0","a=2.0", "a=5.0","a=10.0"]
epsn  = ["\epsilon=0.01, ", "\epsilon=0.1, ", "\epsilon=0.5, ", "\epsilon=1.0, "]
modi = ["k=1", "k=2", "k=3", "k=4", "k=5","k=6","k=7","k=8","k=9","k=10"]

avalues=[0.1,0.2,0.3,0.5,1.0,2.0,5.0,10.0]
```

ENERGIE MEDIE:

```python
plt.figure(figsize=[6,6])

for m, mname in enumerate(mode):
    for a, aname in enumerate(alpha):
        for e, ename in enumerate(eps):
            print('./img_mod/avg'+ename+mname+aname+".png")
            for i in list(range(1,7)):
                t,k = np.loadtxt("./"+ename+"/avgmodes_"+mname+"_t800_"+aname+".dat", usecols=(0, i), unpack=True)
                plt.plot(t[:],k[:],label=modi[i-1])
        
            plt.title(r'$'+epsn[e]+moden[m]+alphan[a]+'$')
            plt.xlabel('t')
            plt.ylabel(r'$ <E_k>_t$')
            plt.legend()
            #plt.show()
            plt.savefig('./img_mod/avg'+ename+mname+aname+".png")
            plt.close()
```

MODI ARMONICI

```python
plt.figure(figsize=[6,6])

for m, mname in enumerate(mode):
    for a, aname in enumerate(alpha):
        for e, ename in enumerate(eps):
            print('./img_mod/mod'+ename+mname+aname+".png")
            for i in list(range(1,7)):
                t,k = np.loadtxt("./"+ename+"/modes_"+mname+"_t800_"+aname+".dat", usecols=(0, i), unpack=True)
                plt.plot(t[:],k[:],label=modi[i-1])
        
            plt.title(r'$'+epsn[e]+moden[m]+alphan[a]+'$')
            plt.xlabel('t')
            plt.ylabel(r'$ <E_k>_t$')
            plt.legend()
            #plt.show()
            plt.savefig('./img_mod/mod'+ename+mname+aname+".png")
            plt.close()
```

POWER SPECTRUM:

```python
plt.figure(figsize=[6,6])
N=10
OmegaM = [0]*N

for i in list(range(N)):
    OmegaM[i]=2*np.sin(np.pi*(i+1)/(2*(N+1)))

for m, mname in enumerate(mode):
    for a, aname in enumerate(alpha):
        for e, ename in enumerate(eps):
            print('./img_mod/ps'+ename+mname+aname+".png")
            t, k = np.loadtxt("./"+ename+"/ps_"+mname+"_t800_"+aname+".dat", usecols=(0, 1), unpack=True)
            plt.plot(t[:],k[:],label='power spectrum')
            plt.scatter(OmegaM[:],[0]*N,label='freq. teoriche')
            plt.title(r'$'+epsn[e]+moden[m]+alphan[a]+'$')
            plt.xlabel('$\omega_k$')
            plt.ylabel(r'$|C(\omega_k)|^2$')
            plt.legend()
            #plt.show()emacs 
            plt.savefig('./img_mod/ps'+ename+mname+aname+".png")
            plt.close()
    
```

ERGODICITY CHECK:

```python
media = [0]*10
devstd = [0]*10
N=10
chiEMA = [[[0]*len(alpha)]*len(mode)]*len(eps)

for e, ename in enumerate(eps):
    for m, mname in enumerate(mode):
        for a, aname in enumerate(alpha):
            for i in list(range(1,11)):
                k = np.loadtxt("./"+ename+"/modes_"+mname+"_t800_"+aname+".dat", usecols=( i), unpack=True)
                lsup = int(len(k))
                linf = int(lsup/2)
                media[i-1]=np.mean(k[linf:lsup-1])
                devstd[i-1]=np.std(k[linf:lsup-1])
            
            mediatot=sum(media)/10
            chi2=0
            for i in list(range(0,10)):
                chi2=chi2+(mediatot-media[i])**2/devstd[i]
            chi2=chi2
            chiEMA[e][m][a]=chi2/(N-1)
            if (chi2<1):
                print('chi2<1 per :',mname,aname,ename)
            
        plt.plot(avalues,chiEMA[e][m][:],label=moden[m])
    plt.hlines(1,0,10)
    plt.yscale('log')
    plt.ylim(bottom=0.01)
    plt.title(r'$'+epsn[e]+'$ test ergodicità')
    plt.xlabel(r'$ \alpha $')
    plt.xscale('log')
    plt.ylabel(r'$ \chi^2 $')
    plt.legend()
    #plt.show()
    plt.savefig('./img_mod/Chi'+ename)
    plt.close()
            
            
```
